```python
import csv

def generate_inline_requests(csv_file_path, data_array=None):
    """
    Generate inline requests from a CSV file or an optional array of indices.

    Args:
        csv_file_path (str): Path to the CSV file.
        data_array (list, optional): List of indices to match with the "No." column in the CSV file.

    Returns:
        list: A list of inline request objects.
    """
    inline_requests = []

    # Read the CSV file into a list of dictionaries
    with open(csv_file_path, "r", encoding="utf-8") as csvfile:
        reader = csv.DictReader(csvfile)
        csv_data = list(reader)

    if data_array:
        # Filter rows based on the indices in data_array
        for index in data_array:
            matching_row = next((row for row in csv_data if int(row.get("No.")) == index), None)
            if matching_row:
                title = matching_row.get("book_title", "Unknown Title")
                description = matching_row.get("description", "No Description Available")

                prompt_text = f"""
                Bạn là một "prompt generator" cho đánh giá sách.
                Viết **một prompt ngắn gọn** để đánh giá cuốn sách "{title}" tập trung vào một **chủ đề nổi bật, thú vị** của câu chuyện. 
                Mô tả ngắn gọn: {description}.
                Chỉ trả về prompt, không giải thích hay liệt kê nhiều ví dụ.

                Ví dụ:
                Tôi muốn viết bài đánh giá về cuốn sách "Hoàng Tử Bé" tập trung vào sự đối lập giữa sự ngây thơ và trí tưởng tượng của trẻ con với sự khô khan, thực dụng của người lớn trong các hành tinh mà Hoàng Tử Bé ghé thăm.

                * Ngắn gọn.
                * Có chủ đề rõ ràng.
                * Không thêm danh sách hay giải thích.
                """

                request_obj = {"contents": [{"parts": [{"text": prompt_text}], "role": "user"}]}
                inline_requests.append(request_obj)
    else:
        # Process all rows in the CSV file
        for row in csv_data:
            title = row.get("book title", "Unknown Title")
            description = row.get("description", "No Description Available")

            prompt_text = f"""
            Bạn là một "prompt generator" cho đánh giá sách.
            Viết **một prompt ngắn gọn** để đánh giá cuốn sách "{title}" tập trung vào một **chủ đề nổi bật, thú vị** của câu chuyện. 
            Mô tả ngắn gọn: {description}.
            Chỉ trả về prompt, không giải thích hay liệt kê nhiều ví dụ.

            Ví dụ:
            Tôi muốn viết bài đánh giá về cuốn sách "Hoàng Tử Bé" tập trung vào sự đối lập giữa sự ngây thơ và trí tưởng tượng của trẻ con với sự khô khan, thực dụng của người lớn trong các hành tinh mà Hoàng Tử Bé ghé thăm.

            * Ngắn gọn.
            * Có chủ đề rõ ràng.
            * Không thêm danh sách hay giải thích.
            """

            request_obj = {"contents": [{"parts": [{"text": prompt_text}], "role": "user"}]}
            inline_requests.append(request_obj)

    return inline_requests

# Example usage
csv_file_path = "book_reviews_prompts.csv"
data_array = [20, 39, 54, 83, 97]  # Example indices to match with the "No." column
inline_requests = generate_inline_requests(csv_file_path, data_array=data_array)
print(inline_requests)
```
